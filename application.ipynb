```python
soup = scraper.get_soup("http://www.anp.gov.br/glossario")
print(soup.prettify())
```

```python
glossary = scraper.create_glossary(soup)
```

```python
resolucoes = []
for doc in glossary.get_sources():
    if doc:
        match = re.match(r"(Resolução ANP nº ([\d]+)(.*?)(\/\d{4}))", doc)
        if match is None:
            continue
        for group in match.groups():
            print(group)
            if group is None:
                continue
        resolucoes.append((match.group(1),match.group(2),match.group(4)))
for resolucao in resolucoes:
    print(resolucao)
```
